 *Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
Найдите их сумму. (на листочке)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)* 


```python
#### 1
import math
### решение для векторов данных в условии
x = [10, 10, 10]
y = [0, 0, -10]
result_vector = []

for i in range (0, len(x)):
    result_vector.append(x[i] + y[i])
print(result_vector)


def length(li):
    length_temp = 0
    for l in li:
        length_temp += l ** 2
    return math.sqrt(length_temp)

print('Длина вектора X =', round(length(x), 2))
print('Длина вектора Y =', round(length(y), 2))
print('Длина результирующего вектора =', round(length(result_vector), 2))

# реализация нахождения длины для произвольного вектора
print('-'*50)
print('Для произвольного вектора')
vector = []
while True:
    s = input('Enter numbers. To exit enter Q.: \n')
    if s.title() == 'Q':
        break
    else:
        try:
            vector.append(float(s))
        except ValueError:
            print('Enter numbers or q for quit')

        sqrnumbers = []
        for el in vector:
            sqrnum = el**2
            sqrnumbers.append(round(sqrnum, 2))
summa = sum(sqrnumbers)

print(f'Your vector is {vector}')
print(sqrnumbers)
print('Length of vector is', round(math.sqrt(summa), 2))
```

```python
def polar(R, phi):
    x = R * np.cos(np.radians(phi))
    y = R * np.sin(np.radians(phi))
    return round(x, 2), round(y, 2)
print(polar(10, 30))
```

Задача 4.2. Решена на листке


*exp(x) + x∙(1 – y) > 1*  
*y = x**2 – 1*


Подставим y из уравнения в неравенство. Получим 2 графика. Экспонента и кубическое уравнение. Экспонента будет больше на трёх участках.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve


fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
x = np.linspace(-5, 5, 300)
y1 = (x**3 - 2 * x + 1)
y2 = np.exp(x)
plt.ylim(-5, 70)
plt.plot(x, y1, color = 'green')
plt.plot(x, y2, color = 'blue')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()


def equations(p):
    x, y = p
    return (y - np.exp(x)), (y - (x**3 - 2 * x + 1))

x1, y1 = fsolve(equations, (-2, 0))
x2, y2 = fsolve(equations, (0, 1))
x3, y3 = fsolve(equations, (4.2, 6))

print(round(x1, 2), round(y1, 2))
print(round(x2, 2), round(y2, 2))
print(round(x3, 2), round(y3, 2))
```
